The "head" should never be more than 2 variables, and the "tail" never more than 2. Since we can pull 50 variables at once this means that we can systematically program in 46 variables for each pull, so that's:
+ dg1: ID's 1 - 46  
+ dg2: ID's 47-92  
+ dg3: ID's 93-138  
+ dg4: ID's 139-184  
+ dg5: ID's 185-230 

```python
dataguide = pd.read_csv('../Data Guide.csv', dtype = str)
dataguide['ID'] = dataguide['ID'].astype(int)
```

```python
dg1 = dataguide[dataguide['ID'].between(1, 46)]
dg2 = dataguide[dataguide['ID'].between(47, 92)]
dg3 = dataguide[dataguide['ID'].between(93, 138)]
dg4 = dataguide[dataguide['ID'].between(139, 184)]
dg5 = dataguide[dataguide['ID'].between(185, 230)]
```

```python
# listparams
head1 = 'NAME'
tail_cols1 = 'StateFIPS'
dflist = [dg1,dg2,dg3,dg4,dg5]
```

```python
# THIS only takes the last two from the big list - save because it works
data_appended = []
for index, element in enumerate(dflist):
    var_list = list(dflist[index]['ACS Variable'])
    #add stuff to variables list
    var_list = deque(var_list)
    var_list.appendleft(head1)
    var_list = list(var_list)
    #make columns list
    col_list = list(dflist[index]['Column Name'])
    #add stuff to columns list
    col_list.append(tail_cols1)
    col_list = deque(col_list)
    col_list.appendleft(head1)
    col_list = list(col_list)
    #API call
    url_str= 'https://api.census.gov/data/2020/acs/acs5?key='+api_key
    predicates= {}
    get_vars= var_list
    predicates["get"]= ",". join(get_vars)
    predicates["for"]= "state:47" #all places
    data= requests.get(url_str, params= predicates)
    col_names = col_list
    data=pd.DataFrame(columns=col_names, data=data.json()[1:], dtype=str)
    data_appended.append(data)
data_appended = pd.concat(data_appended, axis = 1)
data = data_appended
#dflist[index] = data_appended.reset_index(drop = True)
#data_appended = dflist[index-1].merge(dflist[index], how = "outer")
print('Your API call is complete.')
```

```python
def sumlist(x):
    return sum(x)
#function for percent of whole
def percent(x, y):
    try:
        return round((x/y)*100, 2)
    except ZeroDivisionError:
        return 666
```

```python
#% adults 65+ speak language besides English
listadults65overnotenglish = [data['language_65+_spanish'],data['language_65+_indoeuro'],data['language_65+_asianpacificislander'],data['language_65+_other']]
adults65overnotenglish = sumlist(listadults65overnotenglish)
data['%65+speaknotenglish'] = percent(adults65overnotenglish, data['language_65+_total_series'])
#% not English that don't speak English well or at all
listnotwelloratall = [data['language_65+_other_englishnotwell'], data['language_65+_other_englishnotatall'],data['language_65+_spanish_englishnotwell'],
                      data['language_65+_spanish_englishnotatall'],data['language_65+_indoeuro_englishnotwell'], data['language_65_indoeuro_englishnotatall'],
                      data['language_65+_asianpacificislander_englishnotwell'],data['language_65+_asianpacificislander_englishnotatall']]
notwelloratall = sumlist(listnotwelloratall)
data['%65+notenglishdontspeakwelloratall'] = percent(notwelloratall, adults65overnotenglish)
```
