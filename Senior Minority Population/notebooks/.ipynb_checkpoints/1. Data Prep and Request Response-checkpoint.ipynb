## Total and total minority population.

```python
#acs tables
data_appended = []
for i in GNRC:
    url_str= 'https://api.census.gov/data/2020/acs/acs5?key='+api_key
    predicates= {}
    get_vars= ["NAME", "GEO_ID", 'B01003_001E','B03002_003E'] #this is the total population that goes with this data series, it is the same as total pop.
    predicates["get"]= ",". join(get_vars)
    predicates["for"]= "county:{}".format(i)
    predicates["in"]= "state:47"
    data= requests.get(url_str, params= predicates)
    col_names = ['County', 'GEOID','total','nothisp_whitealone','StateFIPS', 'CountyFIPS']
    data=pd.DataFrame(columns=col_names, data=data.json()[1:])
    data['Region'] = 'GNRC'
    data_appended.append(data)
    
data_appended = pd.concat(data_appended)
total = data_appended.reset_index(drop = True)
print('Your API call is complete')
```

```python
total.head(13)
```

```python
cols = ['total', 'nothisp_whitealone']
total[cols] = total[cols].astype(int)
```

```python
total['minority'] = total['total'] - total['nothisp_whitealone']
total['%minority'] = round((total['minority']/total['total'])*100, 2)
```

### Now get counts.

```python
counts = total[['total','nothisp_whitealone','Region']].groupby('Region').sum()
counts
```

```python
minority = counts['total'] - counts['nothisp_whitealone']
minority
```

```python
percminority = (minority/counts['total'])*100
percminority
```

**The total population is 1,962,373, and of that the total minority population is 576,321 (29.37% of the total population). <-- 2019**

**The total population is 1,996,337, and of that the total minority population is 598,698 (29.99% of the total population). <-- 2020**
